```python
train = pd.read_csv("30_Training Dataset_V2/training_data.csv")
```

```python
train.head()
```

```python
print(train.iloc[0,1])
print(train.iloc[0,3])
chang_hand_lst = []
```

```python
search_keyword = "https://community.houseprice.tw/list/"+"台北市"+"_city/"+"水源路"+"_kw/"
# print(search_keyword)
res = requests.get(search_keyword,headers = headers)
res.encoding = 'UTF-8'
soup = BeautifulSoup(res.text,'html.parser')
change_hand = soup.find_all('span', {'class':"txt_color_p"})
# change_hand
```

```python
for i in range(train.shape[0]):
    search_keyword = "https://community.houseprice.tw/list/"+train.iloc[i,1]+"_city/"+train.iloc[i,2]+"_zip/"+train.iloc[i,3]+"_kw/"
    # print(search_keyword)
    res = requests.get(search_keyword,headers = headers)
    res.encoding = 'UTF-8'
    soup = BeautifulSoup(res.text,'html.parser')
    change_hand = soup.find_all('span', {'class':"txt_color_p"})
    print(soup.find_all('span', {'class':"txt_color_p"}))
    # print(change_hand)
    # if len(change_hand) > 0:
    #     print(change_hand[0].text.strip())
    # else:
    #     print("no data")
    chang_hand_lst.append(change_hand[0].text.strip() if len(change_hand) > 0 else "0")
```

```python
train['轉手次數'] = chang_hand_lst
train.head(10)
```
