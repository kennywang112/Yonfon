```python
import pandas as pd
import seaborn as sns #熱力圖
import numpy as np
```

```python
train = pd.read_csv("C:/Users/USER/Desktop/training_data.csv")
```

```python
train.head()
```

```python
import pyproj
```

```python
def tran_coordination(x):
  x1,y1 =x['橫坐標'] ,x['縱坐標']
  proj = pyproj.Transformer.from_crs(3826, 4326, always_xy=True)
  x2, y2 =proj.transform(x1, y1)
  return x2, y2

train[['lon_new','lat_new']]=train.apply(tran_coordination,axis=1,result_type='expand')
```

```python
# ID 不會當作一個判斷
train = train.drop(columns = ['ID','備註','使用分區'])
```

```python
# 和單價比的相關太低
train = train.drop(columns = ['車位個數'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
#分開應變、自變y、x
y_train = train['單價']
x_train = train.drop(columns = ['單價'])
```

```python
# One-Hot Encoding
x_train_encoded = pd.get_dummies(x_train, columns=['縣市','鄉鎮市區','路名','主要用途','主要建材','建物型態'])
```

```python
x_train_encoded
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(x_train_encoded, y_train)
```
