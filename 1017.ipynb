```python
import matplotlib.pyplot as plt
import pandas as pd
from math import cos, sin, pi
```

```python
col_name = ['No Errors','Errors']
small_amount = [50.75,30.77]
medium_amount = [29.85,61.54]
large_amount = [19.4,7.69]
data = {
    col_name[0]: [small_amount[0], medium_amount[0], large_amount[0]],
    col_name[1]: [small_amount[1], medium_amount[1], large_amount[1]]
}
df = pd.DataFrame(data)
df
```

```python
c1 = ['green','yellow','white']
c2 = ['green','yellow','white']
plt.pie(df['Errors'],
        radius=1,
        colors=c1,
        wedgeprops={'linewidth':1,'edgecolor':'black','width':0.2}, autopct='%1.1f%%')
plt.pie(df['No Errors'],
        radius=1.2,
        colors=c2,
        wedgeprops={'linewidth':1,'edgecolor':'black','width':0.2}, autopct='%1.1f%%')

plt.legend(labels=['Small', 'Medium', 'Large'])
plt.title('Invoice Size & Errors\nInner Rung With Errors, OOuter Ring No Errors')

plt.show()
```
